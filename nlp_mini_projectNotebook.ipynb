```python
#importing libraries
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#reading file in other format for implementation of apriori, file has only one 
df = pd.read_csv("/content/sample_data/data.csv", sep = ',',names=["transactions"])
df.head()
```

```python
df = list(df["transactions"].apply(lambda x:x.split(",")))
print(df[0:2]) 
```

```python
#hot encoding the dataset i.e. converting dataset into the form of 0s and 1s
tr = TransactionEncoder()
#fit and transform
tr_arr = tr.fit(df).transform(df)
print(tr_arr)
```

```python
tr.columns_
```

```python
tr_arr = list(tr_arr)
for j in range(len(list(tr_arr))):
  tr_arr[j] = list(tr_arr[j])[1:len(list(tr_arr[j]))]
print(tr_arr[0:5])
```

```python
print(tr.columns_[1:])
print(len(tr_arr[0]))
```

```python
#creating a dataframe out of the nested list
df = pd.DataFrame(tr_arr, columns=tr.columns_[1:])
df
```

Here, the values of the itemsets with support count is displayed, i.e which products appear together a lot in the trasnsactional datset i.e. customers buy these products together many times. eg mineral water is brough many times by customers so shops can give offers on them to increase sales and (french fries, pancakes) are brought together by 2% of all customers so the stores can keep some offers on them

```python
df_apr = apriori(df, min_support = 0.02, use_colnames = True)
df_apr.sort_values(['support'],ascending=False)
```

Here, the lift which suggests both the products are brought together, i.e their corelation, eg, green tea and eggs are brought together and there is some dependence between them

```python
df_ar = association_rules(df_apr, metric="lift", min_threshold=1)
df_ar.sort_values(['lift'],ascending=False)
```

Hre same concept of support items often bought together

```python
df_ar = association_rules(df_apr, metric="support", min_threshold=0.01)
df_ar.sort_values(['support'],ascending=False)
```

Here, the confidence, ie if x is brought then y is definatel brougj os checked, so here, if (spaghetti) -> then	(tomatoes) is brough has 12% confidence, ie if people buy spagetti then they will next go for tomatoes so these roducts can be kept besides each other

```python
df_ar = association_rules(df_apr, metric="confidence", min_threshold=0.05)
df_ar.sort_values(['confidence']).head(50)
```

```python
import pandas as pd
```

```python
#reading file in other format for implementation of apriori, file has only one 
df = pd.read_csv("/content/sample_data/data.csv", sep = ',',names=["transactions"])
df.head()
df = list(df["transactions"].apply(lambda x:x.split(",")))
print(df[0:2]) 
#hot encoding the dataset i.e. converting dataset into the form of 0s and 1s
tr = TransactionEncoder()
#fit and transform
tr_arr = tr.fit(df).transform(df)
print(tr_arr)
tr_arr = list(tr_arr)
for j in range(len(list(tr_arr))):
  tr_arr[j] = list(tr_arr[j])[1:len(list(tr_arr[j]))]
print(tr_arr[0:5])
#creating a dataframe out of the nested list
df = pd.DataFrame(tr_arr, columns=tr.columns_[1:])
df
#Importing Libraries
from mlxtend.frequent_patterns import fpgrowth
#running the fpgrowth algorithm
res=fpgrowth(df,min_support=0.01, use_colnames=True)
# printing top 10
res.head(10)
```

Here, the same data is analyzed with the fp tree algorithm, we can see almost same results

```python
encoding_list=[]
for i in df:
  new_list=[]
  for ini in range(len(li)):
    new_list.append(0)
  for j in i:
    for k in range(len(li)):
      if j==li[k]:
        new_list[k]=1
  encoding_list.append(new_list)

print(encoding_list[0:5])
```
